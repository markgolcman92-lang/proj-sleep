```python
import pandas as pd
import numpy as np

A1 = [80, 77, 76, 71]
B1 = [84, 78, 82, 80]
A2 = [73, 79, 71, 75]
B2 = [83, 82, 81, 84]

all = A1+B1+A2+B2
```

```python
import seaborn as sns

sns.lineplot(all)
```

```python
A = A1+A2
B = B1+B2

result = []
for i in range(100000):
    a_random = np.random.choice(A, size=8, replace=True)
    b_random = np.random.choice(B, size=8, replace=True)

    diff = np.mean(b_random) - np.mean(a_random)
    result.append(diff)

result = np.array(result)

ci_lower = np.percentile(result, 2.5)
ci_upper = np.percentile(result, 97.5)
print(ci_lower, ci_upper)
```

Рекомендации от модели показали статистически значимое улучшение качества сна на +6.5 баллов (95% ДИ: +3.875 до +9.0)

что по итогу

Построена рабочая модель прогноза сна (GAM, R²=0.63, MAE=3.76) на данных браслета

Выявлены ключевые драйверы качества сна: время засыпания/пробуждения, продолжительность сна, дневная активность

 Подтверждена эффективность на A/B тесте: рекомендации улучшают сон на +6.5 баллов (статистически значимо), также для уверенность проведен A/A тест
