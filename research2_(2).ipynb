```python
df[['sleep_duration', 'time2sleep', 'time2wake']] = df[['sleep_duration', 'time2sleep', 'time2wake']].astype(str).applymap(lambda x: int(x.split('.')[0])*60 + int(x.split('.')[1]))
```

```python
def cycle_transform(x):
  x = 2* np.pi * x/1440
  return np.cos(x)
```

```python
df['time2sleep'] = df['time2sleep'].apply(cycle_transform)
```

```python
df['Deep_fraction'] = 100 - df['REM_fraction'] - df['Light_fraction']
```

```python
df['REM_duration'] = df['REM_fraction']* df['sleep_duration']/100
df['Deep_duration'] = df['Deep_fraction']* df['sleep_duration']/100
df['Light_duration'] = df['Light_fraction']* df['sleep_duration']/100
```

```python
from scipy import stats

y = df['sleep_score'].values
n = len(y)

S = 0
for i in range(n-1):
    for j in range(i+1, n):
        S += np.sign(y[j] - y[i])


var_S = (n * (n-1) * (2*n + 5)) / 18

unique, counts = np.unique(y, return_counts=True)
tie_correction = 0
for count in counts:
    if count > 1:
        tie_correction += (count * (count-1) * (2*count + 5))
tie_correction = tie_correction / 18

var_S_corrected = var_S - tie_correction


if S > 0:
    Z = (S - 1) / np.sqrt(var_S_corrected)
elif S < 0:
    Z = (S + 1) / np.sqrt(var_S_corrected)
else:
    Z = 0

p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

tau = (2 * S) / (n * (n - 1))


print(tau)
print(p_value)
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

objective_features = [
    'sleep_duration',
    'Deep_fraction',
    'REM_fraction',
    'Light_fraction',
    'time2sleep',
    'time2wake',
    'kkal'
]

target = 'sleep_detrended'
X = df[objective_features]
y = df[target]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


n_samples = len(X)
n_features = X_scaled.shape[1]
```

```python
n_bootstrap = 2000

coefs_history = np.zeros((n_bootstrap, n_features))


for i in range(n_bootstrap):
    indices = np.random.choice(n_samples, size=n_samples, replace=True)

    X_boot = X_scaled[indices]
    y_boot = y.values[indices]

    model = Ridge(alpha=1.0, random_state=42+i)
    model.fit(X_boot, y_boot)

    coefs_history[i] = model.coef_
```

```python
stats_list = []

for j in range(n_features):
    coefs_for_feature = coefs_history[:, j]

    coef_mean = np.mean(coefs_for_feature)
    coef_median = np.median(coefs_for_feature)
    coef_std = np.std(coefs_for_feature)

    ci_lower = np.percentile(coefs_for_feature, 2.5)
    ci_upper = np.percentile(coefs_for_feature, 97.5)

    significant = (ci_lower > 0) or (ci_upper < 0)

    if abs(coef_mean) > 1e-10:
        coef_cv = coef_std / abs(coef_mean)
    else:
        coef_cv = np.inf

    stability_score = max(0, 1 - coef_cv)

    importance_score = abs(coef_mean) * stability_score

    prob_positive = np.mean(coefs_for_feature > 0)
    prob_negative = np.mean(coefs_for_feature < 0)

    stats_list.append({
        'feature': objective_features[j],
        'coefficient_mean': coef_mean,
        'coefficient_median': coef_median,
        'coefficient_std': coef_std,
        'coefficient_cv': coef_cv,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'significant': significant,
        'stability_score': stability_score,
        'importance_score': importance_score,
        'probability_positive': prob_positive,
        'probability_negative': prob_negative
    })

coef_stats = pd.DataFrame(stats_list)

coef_stats = coef_stats.sort_values('importance_score', ascending=False)
```

```python
np.random.seed(42)
n = len(df)
group_assignments = np.random.choice(['A', 'B'], size=n, replace=True)

group_a = df.loc[group_assignments == 'A', 'sleep_score'].values
group_b = df.loc[group_assignments == 'B', 'sleep_score'].values

print(f"  A: {len(group_a)}")
print(f"  B: {len(group_b)}")

n_iterations = 10000
sample_size = min(len(group_a), len(group_b))

bootstrap_diffs = []

for i in range(n_iterations):
    a_bootstrap = np.random.choice(group_a, size=sample_size, replace=True)
    b_bootstrap = np.random.choice(group_b, size=sample_size, replace=True)

    diff = np.mean(b_bootstrap) - np.mean(a_bootstrap)
    bootstrap_diffs.append(diff)

bootstrap_diffs = np.array(bootstrap_diffs)

ci_90_lower = np.percentile(bootstrap_diffs, 5)
ci_90_upper = np.percentile(bootstrap_diffs, 95)

ci_95_lower = np.percentile(bootstrap_diffs, 2.5)
ci_95_upper = np.percentile(bootstrap_diffs, 97.5)

ci_99_lower = np.percentile(bootstrap_diffs, 0.5)
ci_99_upper = np.percentile(bootstrap_diffs, 99.5)

print("\n ")
print(f"90%  [{ci_90_lower:.3f}, {ci_90_upper:.3f}]")
print(f"95%  [{ci_95_lower:.3f}, {ci_95_upper:.3f}]")
print(f"99%  [{ci_99_lower:.3f}, {ci_99_upper:.3f}]")


observed_diff = np.mean(group_b) - np.mean(group_a)
print(f"\n : {observed_diff:.3f}")


p_value = np.mean(np.abs(bootstrap_diffs) >= np.abs(observed_diff))
print(f"P-value (): {p_value:.4f}")
```

```python
plt.figure(figsize=(8, 5))
n, bins, patches = plt.hist(bootstrap_diffs, bins=30, alpha=0.7, color='lightblue', edgecolor='black', linewidth=0.5)
for patch in patches:
    center = (patch.get_x() + patch.get_x() + patch.get_width()) / 2
    if ci_95_lower <= center <= ci_95_upper:
        patch.set_facecolor('skyblue')
    else:
        patch.set_facecolor('lightgray')

plt.axvline(x=0, color='red', linewidth=2, linestyle='--', label=' ')
plt.axvline(x=observed_diff, color='green', linewidth=2.5, label=f': {observed_diff:.1f}')

plt.xlabel('  sleep_score (B - A)', fontsize=11)
plt.ylabel('', fontsize=11)
plt.title('A/A :    ')

plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```
